```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers

```

```python
(x_train, y_train),(x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#This step was done after running model and getting low accuracy. Let scale and run again. You can skip this step before running model for the first time
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train[0]
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)), #adding flatten layer instead of manufal flattening
    keras.layers.Dense(100, input_shape = (784,), activation ="relu"),
    keras.layers.Dense(20, activation = "relu"),
    keras.layers.Dense(10, activation = "sigmoid")
])
model.compile(optimizer = "adam",
              loss = "sparse_categorical_crossentropy", 
              metrics = ["accuracy"]
)

model.fit(x_train, y_train, epochs = 5)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_predict = model.predict(x_test)
np.argmax(y_predict[4])
y_predict_labels = [np.argmax(i) for i in y_predict]
y_predict_labels[:5]
cm=tf.math.confusion_matrix(labels = y_test, predictions = y_predict_labels)
```

```python
import seaborn as sn
sn.heatmap(cm, annot = True, fmt = "d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
```

***Lets add change activation functions and other parameter to see if it improves the accuracy and reduce wrong predictions***

```python
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)), #adding flatten layer instead of manufal flattening
    keras.layers.Dense(100, input_shape = (784,), activation ="relu"),
    keras.layers.Dense(20, activation = "relu"),
    keras.layers.Dense(10, activation = "softmax")
])

tensor_board = tf.keras.callbacks.TensorBoard(log_dir="logs/generalnn", histogram_freq=1)

model2.compile(optimizer = "adam",
              loss = "sparse_categorical_crossentropy", 
              metrics = ["accuracy"]
)

model2.fit(x_train, y_train, epochs = 5, callbacks = [tensor_board])
```

```python
model2.evaluate(x_test, y_test)
```

```python
y_predict = model2.predict(x_test)
np.argmax(y_predict[4])
y_predict_labels = [np.argmax(i) for i in y_predict]
y_predict_labels[:5]
cm=tf.math.confusion_matrix(labels = y_test, predictions = y_predict_labels)
```

```python
import seaborn as sn
sn.heatmap(cm, annot = True, fmt = "d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
```

***Lets use CNN model to check if it provides better predictive accuracy***

```python

model3 = keras.Sequential([
    keras.layers.Reshape((28, 28, 1), input_shape=(28,28)),
    
    keras.layers.Conv2D(32, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Conv2D(64, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

tensor_board = tf.keras.callbacks.TensorBoard(log_dir="logs/cnn", histogram_freq=1)

model3.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model3.fit(x_train, y_train, epochs=5, callbacks=[tensor_board])
```

```python
model3.evaluate(x_test, y_test)
```

```python
y_predict = model3.predict(x_test)
np.argmax(y_predict[4])
y_predict_labels = [np.argmax(i) for i in y_predict]
y_predict_labels[:5]
cm=tf.math.confusion_matrix(labels = y_test, predictions = y_predict_labels)
```

```python
import seaborn as sn
sn.heatmap(cm, annot = True, fmt = "d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
```
